```python
# 숫자 인덱싱 불가
# 정수INTEGER 1234, 실수FLOAT 0.1 0.3 1.2 
a = 1.9
a = int(a)
print(a)
```

```python
# 문자
# 캐릭터CHARACTER, 문자열STRING 인덱싱 가능
a = '100'
a = int(a)
print(type(a))
```

```python
# 둘리형
a = True
a = False
print(a)

b = False
b = True
print(b)

print(type(a))
print(type(b))
```

```python
# 참조형
# 리스트, 딕셔너리, 셋 , 튜플
# 리스트 먼저
a = [1,2]
print(a) # 리스트를 출력하면 괄호까지 나온다.
print(a[0]) # 리스트의 인덱스번호를 넣어서 따로 출력가능 이게 인덱싱이다. 
b = ['a', '3', ['bac', 22.0]]
print((b[2][0][2]))
```

```python
# 딕셔너리 
# 키값은 인덱싱이 안된다.
a = { 'b' : [1,[5,6],['abv', 'z'],2,3], 'cb' : 333 }
print(a['b'][2][0][1])
```

```python
# 셋이랑 튜풀이랑 같은 취급을한다 # 파이썬한정
# 인덱싱 불가
# 데이터 모음
x = {4, 2, 1, 3, 'ㅁ', 'ㅊ'}
print(x)
```
